#### Matplotlib
- Built on NumPy arrays (and Python)
- Integrates directly with pandas
- Can create basic or advance plots
- Steps
    - Prepare data
    - Create plot(figure)
    - Plot data (axes on figure)
    - Customize plot - lables, etc
    - Save/share plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### Create plot(figure)

```python
plt.plot()
plt.show()
```

#### Plot data (axes on figure)

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
plt.plot(x,y)
plt.show()
```

#### Methods

```python
#1st method
fig = plt.figure() # creates a figure
ax = fig.add_subplot() # add some axes
plt.show()
```

```python
#2nd method
fig = plt.figure() # creates a figure
ax = fig.add_axes([1,1,1,1]) # add some axes
ax.plot(x,y)
plt.show()
```

```python
#3rd method - recommended
fig, ax = plt.subplots() # creates subplots in figure
ax.plot(x,y)
plt.show()
type(fig), type(ax)
```

```python
# anatony
# step 1 - import matplotlib
import matplotlib.pyplot as plt

# step 2 - prepare data
x = [0,2,4]
y = [1,3,5]

# step 3 - setup plot - create figure(canvas) and axes(subplot)
fig, ax = plt.subplots(figsize=(2,2)) # width,height

# step 4 - plot data
ax.plot(x, y)

# step 5 - customize plot
ax.set(title='Demp Plot',
       xlabel='x-axis',
       ylabel='y-axis')

# step 6 - save/show 
# fig.save('image/demo.png')
plt.show()
```

#### Figures with Numpy Arrays

```python
x = np.linspace(0,10,100)
x[:10]
```

```python
# line plot
fig,ax = plt.subplots(figsize=(2,2))
ax.plot(x,x**2)
plt.show()
```

```python
# scatter plot
fig,ax = plt.subplots(figsize=(3,3))
ax.scatter(x,np.sin(x))
plt.show()
```

```python
# plot from dict
inventory = {"apples": 10,"bananas": 5,"oranges": 8}
fig,ax=plt.subplots(figsize=(3,3))
ax.bar(x=inventory.keys(),height=inventory.values(),width=0.5)
ax.set(title='inventory info',ylabel='price()',xlabel='fruits')
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(3,3))
ax.barh(y=inventory.keys(),width=inventory.values(),height=0.5)
ax.set(title='inventory info',xlabel='price()',ylabel='fruits')
plt.show()
```

```python
# histogram
x = np.random.randn(1000)
fig, ax = plt.subplots(figsize=(3,3))
ax.hist(x)
plt.show()
```

### Multi subplots

```python
# option 1 - a figure with multiple axes
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))
```

```python
# option 1 - a figure with multiple axes
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))
# plot to each different axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(x=inventory.keys(),height=inventory.values(),width=0.5)
ax4.hist(np.random.randn(100))
plt.show()
```

```python
# option 2 - a figure with multiple axes
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
```

```python
# option 2 - a figure with multiple axes
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
# plot to each different axes by using index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(x=inventory.keys(),height=inventory.values(),width=0.5)
ax[1,1].hist(np.random.randn(100))
plt.show()
```
