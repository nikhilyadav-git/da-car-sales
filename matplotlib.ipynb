#### Matplotlib
- Built on NumPy arrays (and Python)
- Integrates directly with pandas
- Can create basic or advance plots
- Steps
    - Prepare data
    - Create plot(figure)
    - Plot data (axes on figure)
    - Customize plot - lables, etc
    - Save/share plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### Create plot(figure)

```python
plt.plot()
plt.show()
```

#### Plot data (axes on figure)

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
plt.plot(x,y)
plt.show()
```

#### Methods

```python
#1st method
fig = plt.figure() # creates a figure
ax = fig.add_subplot() # add some axes
plt.show()
```

```python
#2nd method
fig = plt.figure() # creates a figure
ax = fig.add_axes([1,1,1,1]) # add some axes
ax.plot(x,y)
plt.show()
```

```python
#3rd method - recommended
fig, ax = plt.subplots() # creates subplots in figure
ax.plot(x,y)
plt.show()
type(fig), type(ax)
```

```python
# anatony
# step 1 - import matplotlib
import matplotlib.pyplot as plt

# step 2 - prepare data
x = [0,2,4]
y = [1,3,5]

# step 3 - setup plot - create figure(canvas) and axes(subplot)
fig, ax = plt.subplots(figsize=(5,5))

# step 4 - plot data
ax.plot(x, y)

# step 5 - customize plot
ax.set(title='Demp Plot',
       xlabel='x-axis',
       ylabel='y-axis')

# step 6 - save/show 
# fig.save('../image/demo.png')
plt.show()
```
